```python
df = pd.read_csv('Number of households by state Malaysia 2010 - 2020_dataset.csv')
df['Number of households'] = df['Number of households (000)'] * 1000 
df = df[['Country/State','Year','Number of households']]
df.rename(columns={'Country/State':'States'},inplace=True)
df = df[df.States != 'Malaysia']
```

```python
df_2010 = df[df.Year == 2010]
df_2010 = df_2010[['States','Number of households']]
df_2011 = df[df.Year == 2011]
df_2011 = df_2011[['States','Number of households']]
df_2012 = df[df.Year == 2012]
df_2012 = df_2012[['States','Number of households']]
df_2013 = df[df.Year == 2013]
df_2013 = df_2013[['States','Number of households']]
df_2014 = df[df.Year == 2014]
df_2014 = df_2014[['States','Number of households']]
df_2015 = df[df.Year == 2015]
df_2015 = df_2015[['States','Number of households']]
df_2016 = df[df.Year == 2016]
df_2016 = df_2016[['States','Number of households']]
df_2017 = df[df.Year == 2017]
df_2017 = df_2017[['States','Number of households']]
df_2018 = df[df.Year == 2018]
df_2018 = df_2018[['States','Number of households']]
df_2019 = df[df.Year == 2019]
df_2019 = df_2019[['States','Number of households']]
df_2020 = df[df.Year == 2020]
df_2020 = df_2020[['States','Number of households']]
```

```python
df_yearly = pd.merge(df_2010,df_2011, on='States')
df_yearly.rename(columns={'Number of households_x':'2010','Number of households_y':'2011'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2012, on='States')
df_yearly.rename(columns={'Number of households':'2012'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2013, on='States')
df_yearly.rename(columns={'Number of households':'2013'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2014, on='States')
df_yearly.rename(columns={'Number of households':'2014'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2015, on='States')
df_yearly.rename(columns={'Number of households':'2015'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2016, on='States')
df_yearly.rename(columns={'Number of households':'2016'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2017, on='States')
df_yearly.rename(columns={'Number of households':'2017'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2018, on='States')
df_yearly.rename(columns={'Number of households':'2018'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2019, on='States')
df_yearly.rename(columns={'Number of households':'2019'},inplace=True)
df_yearly = pd.merge(df_yearly,df_2020, on='States')
df_yearly.rename(columns={'Number of households':'2020'},inplace=True)
```

```python
df_yearly.loc['Malaysia'] = df_yearly.sum(numeric_only=True, axis=0)
df_yearly['States'] = df_yearly['States'].replace(np.nan, 'Malaysia')
df_yearly['2011 Inc.'] = ((df_yearly['2011'] - df_yearly['2010'])/df_yearly['2010'])*100
df_yearly['2012 Inc.'] = ((df_yearly['2012'] - df_yearly['2011'])/df_yearly['2011'])*100
df_yearly['2013 Inc.'] = ((df_yearly['2013'] - df_yearly['2012'])/df_yearly['2012'])*100
df_yearly['2014 Inc.'] = ((df_yearly['2014'] - df_yearly['2013'])/df_yearly['2013'])*100
df_yearly['2015 Inc.'] = ((df_yearly['2015'] - df_yearly['2014'])/df_yearly['2014'])*100
df_yearly['2016 Inc.'] = ((df_yearly['2016'] - df_yearly['2015'])/df_yearly['2015'])*100
df_yearly['2017 Inc.'] = ((df_yearly['2017'] - df_yearly['2016'])/df_yearly['2016'])*100
df_yearly['2018 Inc.'] = ((df_yearly['2018'] - df_yearly['2017'])/df_yearly['2017'])*100
df_yearly['2019 Inc.'] = ((df_yearly['2019'] - df_yearly['2018'])/df_yearly['2018'])*100
df_yearly['2020 Inc.'] = ((df_yearly['2020'] - df_yearly['2019'])/df_yearly['2019'])*100
df_yearly = df_yearly[['States','2010','2011 Inc.','2011','2012 Inc.','2012','2013 Inc.','2013','2014 Inc.',
                                       '2014','2015 Inc.','2015','2016 Inc.','2016','2017 Inc.','2017','2018 Inc.','2018',
                                       '2019 Inc.','2019','2020 Inc.','2020']]

df_yearly.style.set_caption("Malaysia Household By States").set_table_styles([{
    'selector': 'caption',
    'props': [
        ('color', 'black'),
        ('font-size', '20px'),
        ("text-align", "center"),
        ('text-decoration', 'underline'),
        ('font-family','Arial'),
        ('text-shadow', '2px 2px 5px grey')
    ]},(dict
        (selector='th',props=[('text-align',
                               'left')]))]).format(
    {'2010':'{:,.0f}','2011':'{:,.0f}','2012':'{:,.0f}','2013':'{:,.0f}','2014':'{:,.0f}','2015':'{:,.0f}',
     '2016':'{:,.0f}','2017':'{:,.0f}','2018':'{:,.0f}','2019':'{:,.0f}','2020':'{:,.0f}',
     '2011 Inc.':'{:,.1f}','2012 Inc.':'{:,.1f}','2013 Inc.':'{:,.1f}','2014 Inc.':'{:,.1f}','2015 Inc.':'{:,.1f}',
     '2016 Inc.':'{:,.1f}','2017 Inc.':'{:,.1f}','2018 Inc.':'{:,.1f}','2019 Inc.':'{:,.1f}','2020 Inc.':'{:,.1f}'
    }).set_properties(subset=['States'],**{'text-align': 'left'}).apply(lambda x: ['background: yellow' if x.name in ['Total'] else '' for i in x], axis=1).hide_index()
```

```python
df_mas = df_yearly[df_yearly.States == 'Malaysia']
df_mas = df_mas[['States','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']]
#df_mas = df_mas.reset_index()
#pd.pivot_table(df_mas, index="States")
df_mas = df_mas.T
df_mas = df_mas[df_mas.Malaysia != 'Malaysia']
df_mas = df_mas.reset_index()
df_mas.rename(columns={'index':'Year','Malaysia':'Household Numbers'},inplace=True)
df_mas
```

```python
df_pop = pd.read_csv('Population_by_age_group_sex_and_ethnic_group_2010-2019e_Malaysia.csv')
df_pop['Population'] = df_pop['Population ("000)']*1000
df_pop = df_pop[['Sex','Age Group','Year','Citizen category','Ethnic','Population']]
df_pop = df_pop.groupby('Year').sum()
```

```python
df_new = pd.merge(df_mas,df_pop, on='Year')
df_new['Household (%)'] = (df_new['Household Numbers']/df_new['Population'])*100
df_new['Household Diff'] = df_new['Household Numbers'].diff()
df_new['Population Diff'] = df_new['Population'].diff()
df_new
```
